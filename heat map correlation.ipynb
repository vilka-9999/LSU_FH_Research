```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas
```

```python
# Calculate the correlation matrix
name = 'data_cleaned/split_data/semesters/fall.csv'

data = pandas.read_csv(f'{name}')
data.fillna(0, inplace=True)
#data = data.iloc[:, 1:]
data = data.loc[:, [ 'Session Title', 'Player Name', 'Split Name', 'Tags', 'Duration', 'Distance (miles)', 'Power Plays', 'Energy (kcal)', 'Player Load', 'Top Speed (m/s)', 'Distance Per Min (m/min)', 'Power Score (w/kg)']]
numeric_data = data.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_data.corr()

# Set up the matplotlib figure
plt.figure(figsize=(20, 15))


# Draw the heatmap with the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
```
