```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas
import numpy as np
from sklearn import linear_model
```

```python
name = 'data_cleaned/split_data/semesters/fall.csv'
df = pandas.read_csv(f'{name}')
df = df.loc[:, ['Energy (kcal)', 'Distance Per Min (m/min)']]
```

```python
df.head()
```

```python
df.corr()
```

```python
plt.scatter(x=df['Energy (kcal)'], y=df['Distance Per Min (m/min)'])
```

```python
reg = np.polyfit(df['Energy (kcal)'], df['Distance Per Min (m/min)'], deg=1)
reg
```

```python
trend  = np.polyval(reg, df['Energy (kcal)'])
plt.scatter(x=df['Energy (kcal)'], y=df['Distance Per Min (m/min)'])
plt.scatter(df['Energy (kcal)'], trend)
```

```python
lreg = linear_model.LinearRegression()
lreg.fit(df[['Energy (kcal)']], df['Distance Per Min (m/min)'])
```

```python
plt.scatter(x=df['Energy (kcal)'], y=df['Distance Per Min (m/min)'])
plt.scatter(x=df['Energy (kcal)'], y=lreg.predict(df[['Energy (kcal)']]))
```
